```python
import cv2
import numpy as np

img = cv2.imread('D:/VIT B-TECH ENTC/VIT B-TECH ENTC 6SEM/(2) CV/ZebraCrossing/Images/download.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Enhance contrast with adaptive histogram equalization
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
gray = clahe.apply(gray)

# Threshold the image
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

# Apply morphological operations to clean up the image
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# Apply Gaussian blur to reduce noise
thresh = cv2.GaussianBlur(thresh, (5, 5), 0, 0)

# Detect edges with Canny edge detection
edges = cv2.Canny(thresh, 100, 200, 3)

# Find lines using HoughLinesP
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 40, minLineLength=30, maxLineGap=5)

# Initialize variables for stair detection
y_keeper_for_lines = []
stair_counter = 0
okey = True

# Draw lines and count stairs
out_img = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(out_img, (x1, y1), (x2, y2), (0, 0, 255), 3, cv2.LINE_AA)

    if y_keeper_for_lines:
        for m in y_keeper_for_lines:
            if abs(m - y1) < 15:
                okey = False
                break

    if okey:
        cv2.line(out_img, (0, y1), (img.shape[1], y1), (0, 0, 255), 3, cv2.LINE_AA)
        y_keeper_for_lines.append(y1)
        stair_counter += 1

    okey = True

# Print the number of stairs and show the output
print("Zebra crossing Number:", stair_counter)
cv2.putText(out_img, 'Number: ' + str(stair_counter), (40, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
cv2.imshow('Original Image', img)
cv2.imshow('Processed Image', out_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the image
img = cv2.imread("C:/Users/ADMIN/CV_CP/staircase/CP2.jpeg")

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold the grayscale image to create a binary image
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Find contours in the inverted binary image
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw the detected horizontal contours in the original image
num_contours = 0
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    if h < w and cv2.contourArea(contour) > 60:  # Set a minimum threshold for contour area and keep only horizontal contours
        cv2.drawContours(img, [contour], -1, (0, 0, 255), 2)
        num_contours += 1

# Display the image and print the number of detected contours
cv2.imshow("Contours", img)
print("Number of detected steps:", num_contours)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread("C:/Users/ADMIN/CV_CP/staircase/CP27.jpeg")
cv2.imshow('Source', img)

# Apply Gaussian blur to get good results
img = cv2.GaussianBlur(img, (5, 5), 0, 0)

dst = cv2.Canny(img, 100, 200, 5)
out_img = cv2.cvtColor(dst, cv2.COLOR_GRAY2BGR)
control = cv2.cvtColor(dst, cv2.COLOR_GRAY2BGR)

y_keeper_for_lines = []
lines = cv2.HoughLinesP(dst, 1, np.pi/180, 50, 60, 30)
stair_counter = 0
okey = True
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(control, (x1, y1), (x2, y2), (0, 0, 255), 3, cv2.LINE_AA)

    if y_keeper_for_lines:
        for m in y_keeper_for_lines:
            if abs(m - y1) < 15:
                okey = False
                break

    if okey:
        cv2.line(out_img, (0, y1), (img.shape[1], y1), (0, 0, 255), 3, cv2.LINE_AA)
        y_keeper_for_lines.append(y1)
        stair_counter += 1

    okey = True

cv2.putText(out_img, 'Stair number: ' + str(stair_counter), (40, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
print("Stair Number:", stair_counter)
cv2.imshow('Before', img)
cv2.imshow('Control', control)
cv2.imshow('Detected lines', out_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
